1. **Accuracy** measures the proportion of true results (both true positives and true negatives) in the population.

$$\text{Accuracy} = \frac{TP + TN}{TP + TN + FP + FN}$$
    
2. **Sensitivity** measures the proportion of actual positives that are correctly identified.

$$\text{Sensitivity} = \frac{TP}{TP + FN}$$

3. **Specificity** measures the proportion of actual negatives that are correctly identified.

$$\text{Specificity} = \frac{TN}{TN + FP}$$

4. **Precision** measures the proportion of positive identifications that were actually correct.

$$\text{Precision} = \frac{TP}{TP + FP}$$

### 1.

Accuracy measure the proportion of correct predictions out of all predictions made. One of the examples from it is weather prediction, here if the production line produces a relatively equal proportion of false and correct items, accuracy can be used to evaluate a model that products as accurate or not, also the cost of errors is relatively low and symmetric among this situation.

### 2.

Sensitivity measures the proportion of actual positive cases that the model correctly identifies, which is especially important in situations where missing a positive case to label a negative case as positive. One of the examples is the medical diagnosis for life-threatening diseases. Sensitivity is useful to identify all actual cases to ensure timely treatment, a false negative could result in severe health consequences or even loss of life in this case, whereas a false positive may only lead to additional testing or temporary stress.

### 3.

Specificity measures the proportion of actual negative cases that are correctly identified. It is particularly valuable in scenarios where false positives are more problematic than false negatives, meaning the cost or consequences of misclassifying negatives as positives is high. One of the examples is the legal or criminal investigations, where positive cases are rare, high specificity ensures that people withut the condition are not incorrectly labeld. 

### 4.

Precision measures the proportion of positive predictions that made by the model that are actually correct. It is useful in different situations where false positives carry significant costs or consequences, and we want to ensure that when the model predicts a positive case, it is highly likely to be correct. One of the exampples is hiring situation. The high precision ensures that candidates labeled as suitable are genuinely qualified among this situation.

````python
from sklearn.metrics import confusion_matrix

# Align the test features to the training features
X_test_aligned = ab_reduced_noNaN_test[X_train.columns]  # Ensure feature alignment

# Make predictions using both classifiers
y_pred_clf = clf.predict(X_test_aligned)  # Predictions from clf
y_pred_clf2 = clf2.predict(X_test_aligned)  # Predictions from clf2

# Compute confusion matrices
cm_clf = confusion_matrix(ab_reduced_noNaN_test['Hard_or_Paper'], y_pred_clf)
cm_clf2 = confusion_matrix(ab_reduced_noNaN_test['Hard_or_Paper'], y_pred_clf2)

# Function to calculate sensitivity, specificity, and accuracy
def calculate_metrics(cm):
    TN, FP, FN, TP = cm.ravel()  # Unpack confusion matrix values
    sensitivity = TP / (TP + FN)  # True Positive Rate
    specificity = TN / (TN + FP)  # True Negative Rate
    accuracy = (TP + TN) / cm.sum()  # Overall accuracy
    return sensitivity, specificity, accuracy

# Calculate metrics for clf
sensitivity_clf, specificity_clf, accuracy_clf = calculate_metrics(cm_clf)

# Calculate metrics for clf2
sensitivity_clf2, specificity_clf2, accuracy_clf2 = calculate_metrics(cm_clf2)

# Print results
print("Confusion Matrix for clf:")
print(cm_clf)
print(f"Sensitivity: {sensitivity_clf:.3f}, Specificity: {specificity_clf:.3f}, Accuracy: {accuracy_clf:.3f}\n")

print("Confusion Matrix for clf2:")
print(cm_clf2)
print(f"Sensitivity: {sensitivity_clf2:.3f}, Specificity: {specificity_clf2:.3f}, Accuracy: {accuracy_clf2:.3f}")
````

> The way a **classification decision tree** is fit is that at each step in the construction process of adding a new **decision node splitting rule** to the current **tree structure**, all possible **decision rules** for all possible **predictor variables** are considered, and the combination that improves the **prediction** the most (as measured by the criterion of either "Gini impurity" or "Shannon entropy") and in accordance with the rules of the decision tree (such as the `max_depth` argument) is added to the **classification decision tree**.  Thus overall "criterion" noted above improves with each new **decision node splitting rule**, so the improvement can thus be tracked and the improvement contributions attributed to the **feature** upon which the **decision node splitting rule** is based.  This means the relative contribution of each **predictor variable** to the overall explanatory power of the model can be calculated, and this is what the `.feature_importances_` attribute does. 
>
> Compared to the simplicity of understanding how different **covariates** contribute towards the final **predicted values** of **multiple linear regression models** (by just reading off the equation to see how predictions work), the the complexity of how all the different **features** interact and combine to together to create the final **predictions** from **classification decision trees** can be staggering. But the so-called **feature importance** heuristics allows us to judge how relatively important the overall contributions from different features are in the final decision tree predictions. Now we just need to be sure we're not **overfitting** our **classification decision trees** since they can be so **complex**.
